```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

# Set visualization style
sns.set_style('whitegrid')
```

```python
# Load dataset
df = pd.read_csv('USA_Housing_Cleaned.csv')  # Update with your actual file name

# Display first few rows
df.head()
```

```python
# Check dataset shape
print("Dataset shape:", df.shape)

# Basic info
df.info()

# Summary statistics
df.describe()
```

```python
# Check missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Drop rows with missing values
df = df.dropna()

print("\nAfter dropping missing values:")
print(df.isnull().sum())
```

```python
# Check duplicate rows
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Remove duplicates
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Remove outliers for 'price'
df = remove_outliers_iqr(df, 'price')

print("Shape after outlier removal:", df.shape)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=30, color='blue', alpha=0.7)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('House Price Distribution')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

# Top 10 features most correlated with price
top_features = corr['price'].abs().sort_values(ascending=False).head(10).index

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df[top_features].corr(), annot=True, cmap='coolwarm')
plt.title('Top 10 Features Correlated with Price')
plt.show()
```

```python
sns.pairplot(df[['price', 'sqft_living', 'bedrooms', 'bathrooms']])
plt.show()
```

```python
scaler = MinMaxScaler()

# Select numerical columns to scale
numeric_cols = ['sqft_living', 'sqft_lot', 'bedrooms', 'bathrooms']

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```

```python
print("Available columns:")
print(df.columns.tolist())
```

```python
# Total rooms = bedrooms + bathrooms
df['total_rooms'] = df['bedrooms'] + df['bathrooms']

# Price per square foot
df['price_per_sqft'] = df['price'] / (df['sqft_living'] + 1e-5)  # to avoid division by zero

df.head()
```

```python
# Apply log transformation to 'price'
df['price'] = np.log1p(df['price'])
```

```python
# Features
X = df.drop('price', axis=1)

# Target
y = df['price']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```
